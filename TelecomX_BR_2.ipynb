#ðŸ“Š Carga e anÃ¡lise

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Filter for customers who did not churn
no_churn_df = dados_clientes[dados_clientes['Churn'] == 'No']

# Normalize the customer column to access 'gender' and 'SeniorCitizen'
df_normalizado_customer_no_churn = pd.json_normalize(no_churn_df['customer'])
df_normalizado_customer_no_churn['customerID'] = no_churn_df['customerID']
df_normalizado_customer_no_churn = df_normalizado_customer_no_churn.set_index('customerID')

# Combine the churn information with the normalized customer data
no_churn_combined = no_churn_df[['customerID', 'Churn']].merge(
    df_normalizado_customer_no_churn[['gender', 'SeniorCitizen']],
    left_on='customerID',
    right_index=True
)

# Group by SeniorCitizen and gender and count the occurrences
churn_senior_gender_counts = no_churn_combined.groupby(['SeniorCitizen', 'gender']).size().reset_index(name='count')

# Map SeniorCitizen values to descriptive labels
churn_senior_gender_counts['SeniorCitizen'] = churn_senior_gender_counts['SeniorCitizen'].map({0: 'Not Senior Citizen', 1: 'Senior Citizen'})

# Filter for customers who did churn
churn_df = dados_clientes[dados_clientes['Churn'] == 'Yes']

# Normalize the customer column to access 'gender' and 'SeniorCitizen'
df_normalizado_customer_churn = pd.json_normalize(churn_df['customer'])
df_normalizado_customer_churn['customerID'] = churn_df['customerID']
df_normalizado_customer_churn = df_normalizado_customer_churn.set_index('customerID')

# Combine the churn information with the normalized customer data
churn_combined = churn_df[['customerID', 'Churn']].merge(
    df_normalizado_customer_churn[['gender', 'SeniorCitizen']],
    left_on='customerID',
    right_index=True
)

# Group by SeniorCitizen and gender and count the occurrences for churned customers
churn_senior_gender_counts_yes = churn_combined.groupby(['SeniorCitizen', 'gender']).size().reset_index(name='count')

# Map SeniorCitizen values to descriptive labels for churned customers
churn_senior_gender_counts_yes['SeniorCitizen'] = churn_senior_gender_counts_yes['SeniorCitizen'].map({0: 'Not Senior Citizen', 1: 'Senior Citizen'})


# Create subplots
fig, axes = plt.subplots(1, 2, figsize=(18, 6))

# Plot the "No Churn" chart
ax1 = sns.barplot(x='count', y=churn_senior_gender_counts['SeniorCitizen'] + ' - ' + churn_senior_gender_counts['gender'], data=churn_senior_gender_counts, orient='h', ax=axes[0], palette={'Female': 'coral', 'Male': 'royalblue'}, hue='gender')
ax1.set_title('No Churn')
ax1.set_xlabel('Count')
ax1.set_ylabel('Category')

# Add text labels inside the bars for No Churn plot
for p in ax1.patches:
    width = p.get_width()
    if width > 0: # O grÃ¡fico mostrava um "0" no inÃ­cio da barra e adicionei essa informaÃ§Ã£o para remover
        ax1.text(width/2,
                 p.get_y() + p.get_height()/2,
                 '{:1.0f}'.format(width),
                 ha='center',
                 va='center',
                 color='white',
                 fontweight='bold')

# Plot the "Churn" chart
ax2 = sns.barplot(x='count', y=churn_senior_gender_counts_yes['SeniorCitizen'] + ' - ' + churn_senior_gender_counts_yes['gender'], data=churn_senior_gender_counts_yes, orient='h', ax=axes[1], palette={'Female': 'coral', 'Male': 'royalblue'}, hue='gender')
ax2.set_title('Churn')
ax2.set_xlabel('Count')
ax2.set_ylabel('Category')

# Add text labels inside the bars for Churn plot
for p in ax2.patches:
    width = p.get_width()
    if width > 0: # O grÃ¡fico mostrava um "0" no inÃ­cio da barra e adicionei essa informaÃ§Ã£o para remover
        ax2.text(width/2,
                 p.get_y() + p.get_height()/2,
                 '{:1.0f}'.format(width),
                 ha='center',
                 va='center',
                 color='white',
                 fontweight='bold')

plt.tight_layout() # Adjust layout to prevent overlapping titles
plt.show()
```
